To give each portfolio and good and fair benchmark I decided to compare them both to the S&P 500. According to Wikipedia, "This fund is the largest ETF in the world." Major hedge funds and banks use the S&Y 500 as a very solid basis to help them determine where the overall market is headed. If the S&Y 500 is preforming poorly, traders know that the weeks ahead might be bumping. But if the S&PY 500 is going up, traders are more relaxed as they have more "wiggle" room to play with. Because of this, it is also a great basis to comapare both portfolios too.

"The S&P 500, or simply the S&P, is a stock market index that measures the stock performance of 500 large companies listed on stock exchanges in the United States.
```python
# Pull SPY500 from Yahoo finance 
def SPY_Benchmark():
    start = datetime.datetime(2010, 5, 12)
    end = datetime.datetime.now()

    df = web.DataReader("SPY", "yahoo", start, end) 

    df = df[['Adj Close']]
    df.rename(columns={'Adj Close':'Value'}, inplace=True)

    df["Value"] = (df["Value"]-df["Value"][0]) / df["Value"][0] * 100.0
    return df
```

Now we can do the following to compare each Portfolio to the S&P 500

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))

# Brings in the data to our data frames
Portfolio_1 = pd.read_pickle('Portfolio_1.pickle')
# Portfolio_2 = pd.read_pickle('Portfolio_2.pickle')
benchmark = SPY_Benchmark() # Calls our newly created function

# Plots the data from the Two data frames onto one graph
Portfolio_1.plot(ax=ax1, linewidth=1)
# Portfolio_2.plot(ax=ax1)
benchmark.plot(color='k',ax=ax1, linewidth=4) # The plot was modified so that it can be easily identified

# Shows the plotted data
#plt.legend().remove()
plt.show()
```
